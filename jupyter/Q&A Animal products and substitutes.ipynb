# **Question (Q)**: Chicken, pork or beef?
**Short answer (SA)**: Chicken. Pork is the second best option. All meat products have a high environmental impact, but the impact of beef is almost three times higher than that of chicken or pork. Consider eating less meat: this is good for the climate because around 5 kg of plant-based feed is needed to produce 1 kg of meat.<br>
**Extended answer (EA)**: According to CE Delft, the price of beef should be around 40% higher if the damage to the environment would have been calculated in its price. Beef industry produces large amounts of manure and ammonia emissions and leads to deterioration of air- and water quality. For example, a weekly meal of steak for a family of 4 people produces 700 kilos of CO2 emissions on an annual basis. To compensate for this, 35 trees have to grow for a year. Also for human health eating red meat is not beneficial. These are some tips to reduce beef consumption:
1. Consume more legumes and nuts - this is the best source of protein climate-wise.
2. To start out, introduce days without meat in your weekly routine. Every meal matters!
3. If you still prefer meat, go for chicken. 
4. If you still prefer red meat, better buy processed beef like sausages and minced meat than steaks. Processed meat comes from dairy cows, and thus is more environmentally friendly. Meat is a side product of dairy cows, whereas steaks come from cows grown for meat only.<br>
**Sources (S)**:<br>
https://ce.nl/publicaties/vuile-handen-een-vingerafdruk-van-nederland-in-de-mondiale-economie/ <br>
https://www.milieucentraal.nl/eten-en-drinken/milieubewust-eten/vlees-vis-of-vega/#de-impact-van-keuzes-op-je-bord

```python
import matplotlib.pyplot as plt

# kg CO2 equivalent per kg of meat
meat_emissions = {
    'beef': 16.6, 
    'chicken': 6.3,
    'pork': 6.7
}

colors = ['#FF91AA', '#F8A036', '#5DACE2']

fig, ax = plt.subplots()

ax.pie(meat_emissions.values(), explode = [0, 0.1, 0.1], labels = meat_emissions.keys(), colors = colors, startangle = 90)
ax.set_title('CO2e emissions per kilogram of meat')

plt.savefig("plots/chicken_pork_beef.png", dpi=300)
plt.show()

```

# **Q:** Which cow milk substitute is better? / Soy, oat, rice or almond milk?
**SA:** Oat. According to a recent Oxford study, oat milk is the most sustainable option when it comes to milk substitutes. If we look at greenhouse gases emissions, all four drinks score around three times better than cow's milk. Even though almond milk shows the lowest emissions, its production is still very harmful to the environment. <br>
**EA:** The choice of the most sustainable dairy alternative is not straightforward. One should look not only at CO2e emissions but also at environmental side effects, and of course at the nutritional value. Rice milk scores low in terms of the latter two: little nutrition, a lot of CO2e. But when it comes to the production drawbacks, almond is the least attractive option. Production of one glass of almond milk leads to only 140g CO2e emissions but requires almost 74 liter of water, more than any other milk substitute. This means that almond farming, which is mainly located in California, US, has a drought effect on the environment, not to mention the high pressure on honeybees.<br>
Though soy production is exhausting for the environment as well because it causes deforestation, its main use is cattle feed and not human consumption. In other words, soy drink is a sustainable alternative to milk as it cuts the food chain in half. It is important to note though that consuming enough dairy products every day is still very important. Soy drink does not have the same health effects as dairy, but often contains added vitamin B12 and calcium. The nutritional value of oat drink compares to soy, and there are no environmental consequences associated with its consumption, even taken into account its increasing popularity: the Oxford study declared it the winner in the battle of plant-based drinks.<br>
The Guardian expects another major player to join the battle soon: hazelnuts. In the meantime, you can give oat and/or soy drink a try by putting them in your porridge, pancake batter, coffee or tea. It might take some time to get used to the new taste, but even if it still does not taste good after a while, do not worry - new alternatives are coming!<br>
**S:** <br>
https://books.google.com.ar/books?id=Pym9DwAAQBAJ<br>
“Reducing food’s environmental impacts through producers and consumers” by J. Poore and T. Nemecek (2018)

```python
# Nutriscore is based on cronometer.com data and is a percentage of daily nutritional targets met with 200 ml of milk,
# calculated for a healthy middle-sized adult

# type of milk : [g CO2e per serving (200ml), nutriscore]
milk_emissions_nutriscore = {
    'cow milk': [634, 14],
    'oat milk': [196, 12],
    'soy milk': [200, 13],
    'rice milk': [229, 8],
    'almond milk': [140, 4]
}

colors = ['#FF91AA', '#AE84D3', '#80862C', '#5DACE2', '#F8A036']

fig, (emissions, nutriscore) = plt.subplots(1,2)

emissions.barh(list(milk_emissions_nutriscore.keys()), 
              list(item[0] for item in milk_emissions_nutriscore.values()),
              color = colors)
emissions.set_title('Grams CO2e per cup')

nutriscore.barh(list(item[0] for item in milk_emissions_nutriscore.keys()),
               list(item[1] for item in milk_emissions_nutriscore.values()),
               color = colors)
nutriscore.set_title('Nutritional value')

fig.suptitle('CO2e emissions and nutritional value of\nmilk substitutes compared to cow milk', y = 1.1, fontsize = 14)
plt.savefig('plots/milk_substitutes.png', dpi=300)
plt.show()
```
