# Multiple plots in one figure, suptitle

```python
import matplotlib.pyplot as plt


# Create multiple axes within one figure

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (10,5))
fig.suptitle('Main title', fontsize = 16)
```

# Line plot + custom colors

```python
# When plotting from two lists, they must be the same size


# Monthly average precipitation
boulder_monthly_precip = [0.70, 0.75, 1.85, 2.93, 3.05, 2.02, 
                          1.93, 1.62, 1.84, 1.31, 1.39, 0.84]

# Month names for plotting
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", 
          "Aug", "Sept", "Oct", "Nov", "Dec"]

fig, ax = plt.subplots()
ax.plot(months, boulder_monthly_precip, color = 'cyan') # https://matplotlib.org/stable/gallery/color/named_colors.html
plt.show() # to avoid "[<matplotlib.lines.Line2D at 0x22c2ce68790>]" in the output
```

# Scatter plot + custom markers

```python
fig, ax = plt.subplots()
ax.scatter([1,3,4,5,6,3], [5,6,7,2,3,4], marker = 'D') # all markers: https://matplotlib.org/1.4.1/api/markers_api.html
```

# Bar plot + set title and axes labels, edgecolor, transparency

```python
fig, ax = plt.subplots()
ax.bar(['apples', 'bananas', 'strawberries'], 
       [3,5,2], 
       color = 'slateblue', 
       edgecolor = 'black',
       alpha = 0.6) # 0.0 is completely transparent

ax.set(title = 'Unavoidable waste',
       xlabel = 'fruit',
       ylabel = 'percentage\nwasted')

#rotate labels
plt.setp(ax.get_xticklabels(), rotation = 45)

plt.show()
```

# Save a Figure as a file

```python
fig, ax = plt.subplots()
ax.barh(['a', 'b', 'c'], [4,6,8])
plt.savefig("plots/testplot.png") # saves the last rendered figure
```
